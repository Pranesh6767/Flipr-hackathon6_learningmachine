### Dealing with rows containing empty values

```python
raw_data2 = raw_data.drop(['City','State','Type','SWM'],axis = 1, inplace = False)
```

```python
raw_data2
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

imp = IterativeImputer(max_iter=20, random_state=0)
imp.fit(raw_data2)
raw_data3 = imp.transform(raw_data2)
```

```python
raw_data4 = pd.DataFrame(data=raw_data3, columns=raw_data2.columns)
raw_data4
```

```python
raw_data4['City'] = raw_data['City']
```

```python
raw_data4['State'] = raw_data['State']
raw_data4['Type'] = raw_data['Type']
raw_data4['SWM'] = raw_data['SWM']
```

```python
raw_data4
```

```python
raw_data4.columns
```

```python
raw_data4 = raw_data4[['City', 'State', 'Type', 'SWM','Population [2011]', 'Sex Ratio', 'Median Age', 'Avg Temp',
       'Toilets Avl', '# of hospitals', 'Foreign Visitors', 'Covid Cases']]
```

```python
raw_data4
```

```python
missing_data = raw_data4.isnull()
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")
```

```python
raw_data4.dropna(subset=["SWM"], axis=0, inplace=True)
```

```python
cleaned_data = raw_data4.copy()
```

```python
len(cleaned_data)
```

```python
cleaned_data.to_excel('cleaned_data.xlsx')
```

```python
cleaned_data.dtypes
```

## Data normalization

```python
# replace (original value) by (original value)/(maximum value)
print('Population [2011]: ',cleaned_data['Population [2011]'].max())
cleaned_data['Population [2011]'] = cleaned_data['Population [2011]']/cleaned_data['Population [2011]'].max()

print('Factor Sex Ratio ',cleaned_data['Sex Ratio'].max())
cleaned_data['Sex Ratio'] = cleaned_data['Sex Ratio']/cleaned_data['Sex Ratio'].max()

print('Factor Median Age: ',cleaned_data['Median Age'].max())
cleaned_data['Median Age'] = cleaned_data['Median Age']/cleaned_data['Median Age'].max()

print('Avg Temp: ',cleaned_data['Avg Temp'].max())
cleaned_data['Avg Temp'] = cleaned_data['Avg Temp']/cleaned_data['Avg Temp'].max()

print('Toilets Avl: ',cleaned_data['Toilets Avl'].max())
cleaned_data['Toilets Avl'] = cleaned_data['Toilets Avl']/cleaned_data['Toilets Avl'].max()

print('# of hospitals: ',cleaned_data['# of hospitals'].max())
cleaned_data['# of hospitals'] = cleaned_data['# of hospitals']/cleaned_data['# of hospitals'].max()
```

```python
print('Foreign Visitors: ',cleaned_data['Foreign Visitors'].max())
cleaned_data['Foreign Visitors'] = cleaned_data['Foreign Visitors']/cleaned_data['Foreign Visitors'].max()
```

## Encoding categorical data

```python
dummy_variable_1 = pd.get_dummies(normal_data["SWM"])
dummy_variable_1.head(10)
```

```python
dummy_variable_1.rename(columns={'HIGH':'SWM-HIGH','LOW':'SWM-LOW','MEDIUM':'SWM-MEDIUM'}, inplace=True)
dummy_variable_1.head()
```

```python
cleaned_final_data = pd.concat([normal_data, dummy_variable_1], axis=1)
```
