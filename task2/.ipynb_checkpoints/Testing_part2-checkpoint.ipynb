```python
raw_data = pd.read_excel('test_data_cleaned.xlsx')
```

```python
raw_data.drop(['Unnamed: 0'],axis = 1, inplace = True)
raw_data
```

```python
cleaned_data = raw_data.copy()
```

```python
# replace (original value) by (original value)/(maximum value)
cleaned_data['Population [2011]'] = cleaned_data['Population [2011]']/12442373.0

cleaned_data['Sex Ratio'] = cleaned_data['Sex Ratio']/1042.0

cleaned_data['Median Age'] = cleaned_data['Median Age']/32.0

cleaned_data['Avg Temp'] = cleaned_data['Avg Temp']/40.0

cleaned_data['Toilets Avl'] = cleaned_data['Toilets Avl']/227.0

cleaned_data['# of hospitals'] = cleaned_data['# of hospitals']/159.0

cleaned_data['Foreign Visitors'] = cleaned_data['Foreign Visitors']/4684707.0
```

```python
cleaned_data
```

```python
dummy_variable_1 = pd.get_dummies(cleaned_data["SWM"])
dummy_variable_1.rename(columns={'HIGH':'SWM-HIGH','LOW':'SWM-LOW','MEDIUM':'SWM-MEDIUM'}, inplace=True)
cleaned_final_data = pd.concat([cleaned_data, dummy_variable_1], axis=1)
cleaned_final_data
```

```python
dummy_variable_1 = pd.get_dummies(cleaned_final_data["Type"])
dummy_variable_1.head(10)
```
