```python
import pandas as pd
```

```python
raw_data = pd.read_excel('Test_dataset.xlsx',sheet_name = 1)
raw_data
```

```python

raw_data2 = raw_data.drop(['City'],axis = 1, inplace = False)



from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

imp = IterativeImputer(max_iter=20, random_state=0)
imp.fit(raw_data2)
raw_data3 = imp.transform(raw_data2)
```

```python
raw_data4 = pd.DataFrame(data=raw_data3, columns=raw_data2.columns)
raw_data4
```

```python
raw_data4['City'] = raw_data['City']
```

```python
missing_data = raw_data4.isnull()
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")
```

```python
raw_data4.to_excel('TSP_cleaned_data.xlsx')
```
